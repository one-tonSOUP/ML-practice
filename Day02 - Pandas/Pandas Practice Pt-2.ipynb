### bfill and ffill arguments in Fillna
`ffill` is forward fill, check the previous row value and update it.
`bffil` is backward fill, check the next row value and update it

```python
check_ffill=pd.DataFrame(
            {
                "ColOne" : [6, None, None, 10, 5],
                "ColTwo" : [None, 7, 2, 3, None],
                "ColThree" : [4, None, 8, 5, None],
                "ColFour" : [5, 4, 2, None, 0]
            })
```

```python
check_ffill
```

```python
check_ffill.fillna(method = 'ffill')
```

```python
check_ffill.fillna(method = 'bfill')
```

### Groupby

```python
df = pd.DataFrame(
    {
        'Anime/Manga' : ['One Piece', 'One Piece', 'AOT', 'AOT','Gintama','Gintama'],
        'Episodes/Chapter' : [1015, 1047, 75, 129, 366, 704]
    })
```

```python
df
```

```python
df.groupby(['Anime/Manga']).mean()
```

```python
df.groupby(['Anime/Manga']).max()
```

```python
df.groupby(['Anime/Manga']).agg(['mean', 'max', 'min', 'count'])
```
