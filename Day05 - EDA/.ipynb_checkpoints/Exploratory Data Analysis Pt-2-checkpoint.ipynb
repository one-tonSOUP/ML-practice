### CountPlot for Primary & Secondary Types of Ability

```python
ability_types = ['Primary Type', 'Secondary type']
for type in ability_types:
    plt.figure(figsize = (20, 10))
    sns.countplot(x = dataset[type],
                  order = dataset[type].value_counts().index,
              palette = sns.color_palette("husl", dataset[type].nunique()));
    plt.title(f"\n{type} of Ability\n", fontsize = 25, fontweight = 'bold')
    plt.xlabel(type, fontsize = 15)
    plt.ylabel("Count", fontsize = 15)
    plt.xticks(rotation = 45, fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.show()
```

## Looking at the top performing Pokemons in terms of the Features.

```python
print(f"{BOLD}Unique Pokemons -{RESET}")
print(f"Unique Names found             : {BOLD}{dataset['Name'].nunique()}{RESET}")
print(f"Unique Secondary Names found   : {BOLD}{dataset['Name2'].nunique()}{RESET}")
```

```python
features = ['Attack', 'Defense', 'HP', 'Sp.Attack', 'Sp.Defense', 'Speed']

for feature in features:
    max_idx = dataset[feature].idxmax()
    min_idx = dataset[feature].idxmin()

    max_pokemon = dataset.loc[max_idx, ["Name", feature]]
    min_pokemon = dataset.loc[min_idx, ["Name", feature]]

    print(f"{BOLD}{feature}{RESET}:")
    print(f"\tMax → {BOLD}{max_pokemon['Name']}{RESET} ({max_pokemon[feature]})")
    print(f"\tMin → {BOLD}{min_pokemon['Name']}{RESET} ({min_pokemon[feature]})\n")
```

```python
def plot_rank(n, features = ['Attack', 'Defense', 'HP', 'Sp.Attack', 'Sp.Defense', 'Speed', 'Total']):
    for feature in features:
        plt.figure(figsize = (20, 10))
        top_n = dataset.nlargest(n, feature).sort_values(feature, ascending = False)
        sns.barplot(x = 'Name',
                    y = feature,
                    data = top_n,
                    palette = 'viridis')
        plt.title(f"\n\nRanking of Pokemon in terms of `{feature}`.\n", fontsize = 25, fontweight = 'bold')
        plt.xlabel("\n" + feature, fontsize = 15)
        plt.ylabel("Count", fontsize = 15)
        plt.xticks(rotation = 45, fontsize = 20)
        plt.yticks(fontsize = 20)
        plt.show()
```

```python
#features = ['Attack', 'Defense', 'HP', 'Sp.Attack', 'Sp.Defense', 'Speed']
plot_rank(10)
```
