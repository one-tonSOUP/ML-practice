Here `M` means **Malignant** & `B` means **Benign**

```python
data.drop(['Unnamed: 32', 'id'], axis = 1, inplace = True)
```

```python
X = data.drop('diagnosis', axis = 1)
Y = data['diagnosis']

malignant = data[data.diagnosis == 'M']
benign = data[data.diagnosis == 'B']
```

```python
plt.figure(figsize = (15, 8))

plt.subplot(2, 2, 1)
plt.scatter(malignant.radius_mean, malignant.texture_mean, color = 'red')
plt.scatter(benign.radius_mean, benign.texture_mean, color = 'blue')
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")

plt.subplot(2, 2, 2)
plt.scatter(malignant.smoothness_mean, malignant.compactness_mean, color = 'green')
plt.scatter(benign.smoothness_mean, benign.compactness_mean, color = 'yellow')
plt.xlabel("Smooth Mean")
plt.ylabel("Compact Mean")

plt.subplot(2, 2, 3)
plt.scatter(malignant.texture_worst, malignant.radius_worst, color = 'purple')
plt.scatter(benign.texture_worst, benign.radius_worst, color = 'pink')
plt.xlabel("Texture Worst")
plt.ylabel("Radius Worst")

plt.subplot(2, 2, 4)
plt.scatter(malignant.smoothness_worst, malignant.compactness_worst, color = 'black')
plt.scatter(benign.smoothness_worst, benign.compactness_worst, color = 'grey')
plt.xlabel("Smooth Worst")
plt.ylabel("Compactness Worst")

plt.show()
```

## Scaling & Encoding

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
minmax = MinMaxScaler()
le = LabelEncoder()

x = minmax.fit_transform(X)
y = le.fit_transform(Y)
```

## Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.22, random_state = 42)
```

## Training the Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 4)
model.fit(X_train, y_train)
predict = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
