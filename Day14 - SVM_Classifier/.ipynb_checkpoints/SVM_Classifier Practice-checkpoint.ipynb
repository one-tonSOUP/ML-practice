### Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
data.nunique()
```

```python
# Separating Categorical and Numerical Features ..

cat, num = list(), list()

for col in data.columns:
    if data[col].dtype == 'O':
        cat.append(col)
    else:
        num.append(col)
```

```python
print(f"Categorical features found : {cat}\n\nNumerical features found : {num}")
```

```python
le = LabelEncoder()
```

```python
for category in cat:
    data[category] = le.fit_transform(data[category])
```

```python
data.head(5)
```

```python
data.info()
```

### Feature Selection

```python
from sklearn.feature_selection import chi2, mutual_info_classif, f_classif, VarianceThreshold
from sklearn.feature_selection import SelectKBest

from sklearn.model_selection import train_test_split

from sklearn.svm import SVC
```

```python
varThresh = VarianceThreshold()
transform_data = varThresh.fit_transform(data)
```

```python
print(transform_data.shape)
print(data.shape)
```

```python
X = data.drop('salary', axis = 1)
y = data.salary
    
minmax = MinMaxScaler()
X = minmax.fit_transform(X)

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.78, random_state = 42)

model = SVC(kernel = 'linear')

selectK = SelectKBest(chi2, k = 5)
selectK.fit(x_train, y_train)
x_trans = selectK.transform(x_train)
```

```python
x_trans.shape
```

```python
y_train.shape
```

```python
model = model.fit(x_train, y_train)

predict = model.predict(x_test)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
