### Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
data.nunique()
```

```python
# Separating Categorical and Numerical Features ..

cat, num = list(), list()

for col in data.columns:
    if data[col].dtype == 'O':
        cat.append(col)
    else:
        num.append(col)
```

```python
print(f"Categorical features found : {cat}\n\nNumerical features found : {num}")
```

```python
le = LabelEncoder()
```

```python
for category in cat:
    data[category] = le.fit_transform(data[category])
```
