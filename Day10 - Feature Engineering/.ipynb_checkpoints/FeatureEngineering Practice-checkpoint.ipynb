# Feature Selection
Feature selection is primarily focused on removing non-informative or redundant predictors from the model. When you are done creating hundreds of thousands of features, itâ€™s time to select a few of them. Well, we should never create hundreds of thousands of useless features. Having too many features pose a problem well known as the curse of dimensionality. If you have a lot of features, you must also have a lot of training samples to capture all the features.

### Step for forwarding Selection
 - Start with the empty feature set
 - Try the remaining feature
 - Estimate classification/regression error for adding each feature in the model
 - Select a feature that gives maximum improvement
 - Stop when there is no significant improvement
### Step for Backward Selection
 - Start with a complete feature set
 - Try the remaining feature
 - Estimate classification/regression error for adding each feature in the model
 - Drop feature that gives less improvement
 - Stop when there is no significant improvement
### Methods/Technique
#### Univariate
 - Pearson Correlation
 - F-score
 - Chi-square
 - Signal to noise ratio
#### Multivariate
 - Compute 'w' on all features
 - Remove feature with smallest 'w'
 - Recompute 'w' on reduced data

## Data Preprocessing

```python
data['salary'] = data['salary'].apply(lambda x: 1 if '>50K' in x else 0)
```

```python
data_encoded = pd.get_dummies(data.drop('salary', axis=1), drop_first=True)
data_encoded['salary'] = data['salary']
```

```python
corr_matrix = data_encoded.corr()
```

```python
target_corr = corr_matrix['salary'].sort_values(ascending=False)
print("Top correlations with salary - \n")
for feature, corr in target_corr[:10].items():
    print(f"{feature:>35} : {B}{corr:.2f}{R}")
```

```python
plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title("Full Feature Correlation Matrix")
plt.show()
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(data_encoded)
```

```python
X = pd.DataFrame(X_scaled, columns=data_encoded.columns)
y = data['salary']
```

```python
# Evaluation function
def evaluate_model(X_selected, y, title):
    X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.25, random_state=42)
    model = LogisticRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)

    print(f"\n{B}{title}{R} Evaluation:")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"ROC AUC Score: {roc_auc:.4f}")

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {title}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Univariate Selection (F-Score)

```python
f_selector = SelectKBest(score_func=f_classif, k=5)
X_f = f_selector.fit_transform(X, y)
selected_f = X.columns[f_selector.get_support()]
print("Selected features (F-score):", selected_f.tolist())
evaluate_model(X[selected_f], y, title="F-score Selection")
```

### Uniariate Selection (Chi-Square)

```python
X_scaled = MinMaxScaler().fit_transform(X)
chi_selector = SelectKBest(score_func=chi2, k=5)
X_chi = chi_selector.fit_transform(X_scaled, y)
selected_chi = X.columns[chi_selector.get_support()]
print("Selected features (Chi-square):", selected_chi.tolist())
evaluate_model(X[selected_chi], y, title="Chi-square Selection")
```

### Recursive Feature Elimination

```python
model = LogisticRegression()
rfe = RFE(model, n_features_to_select=5)
rfe.fit(X, y)
selected_rfe = X.columns[rfe.support_]
print("Selected features (RFE):", selected_rfe.tolist())
evaluate_model(X[selected_rfe], y, title="RFE Selection")
```

### Feature Importance via Weights

```python
model = LogisticRegression()
model.fit(X, y)
importance = np.abs(model.coef_[0])
top_features = X.columns[np.argsort(importance)[-5:]]
print("Top 5 features by weight:", top_features.tolist())
evaluate_model(X[top_features], y, title="Weight-Based Selection")
```

### Correlation Filter (Pearson)

```python
correlations = X.corrwith(y).abs()
selected_corr = correlations[correlations > 0.1].index.tolist()
print("Selected features (Pearson > 0.1):", selected_corr)
evaluate_model(X[selected_corr], y, title="Pearson Correlation")
```
