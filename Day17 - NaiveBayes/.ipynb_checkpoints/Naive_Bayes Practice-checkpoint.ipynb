#### Detecting dataset encoding:

```python
import chardet

with open("spam.csv", 'rb') as f:
    raw_data = f.read()
    result = chardet.detect(raw_data)
    detected_encoding = result['encoding']
    print(f"Detected encoding : {B}{detected_encoding}{R}")

data = pd.read_csv("spam.csv", encoding = detected_encoding)
print(f"{I}Dataset has been read successfully ..{R}")
```

### Check the Common Words in Text

```python
from collections import Counter

common_counter = Counter(" ".join(data[data['v1'] == 'ham']['v2']).split()).most_common(20)
ham = pd.DataFrame(common_counter)
ham = ham.rename(columns = {0: "Common words in HAM", 1 : 'count'})
common_counter = Counter(" ".join(data[data['v1'] == 'spam']['v2']).split()).most_common(20)
spam = pd.DataFrame(common_counter)
spam = spam.rename(columns = {0: "Common words in SPAM", 1 : 'count'})
```

### Feature Selection

Text preprocessing, tokenization and filtering of common words/stopwords can be transformed into Vectors.

Stopwords is nothing but common words.

```python
token = text.CountVectorizer(stop_words = 'english')
X = token.fit_transform(data['v2'])
np.shape(X)
```

```python
le = LabelEncoder()
Y = le.fit_transform(data['v1'])
```

```python
sns.countplot(x = Y, data = data, palette = ['green', 'red'])
plt.show()
```

### **0**  :  **`HAM`**
### **1**  :  **`SPAM`**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

## Multinomial Naive Bayes Classifier
---
The Multinomial Naive Bayes algorithm is a Bayesian learning approach popular in Natural Language Processing(NLP). The program guesses the tag of a text, such as an email or a neewspaper story, using the Bayes theorem. It calculates each tag's likelihood for a given sample and outputs the tag with the greatest chance.

It is a classification algorithm that is suitable with discrete featrues(e.g., word counts for text classification)

```python
model_mnbc = MultinomialNB(alpha = 0.6)
model_mnbc.fit(X_train, y_train)

predict_mnbc = model_mnbc.predict(X_test)

print(f"Classification Report : \n{B}{Red}{classification_report(y_true = y_test, y_pred = predict_mnbc)}{R}")
print("- " * 30)
print(f"Train Score  : {B}{Blue}{model_mnbc.score(X_train, y_train)}{R}")
print(f"Test Score   : {B}{Blue}{model_mnbc.score(X_test, y_test)}{R}")
```

## Gaussian Naive Bayes Classifier
---
Gaussian Naive Bayes suppoorts continuous valued features and models each as conforming to a Gaussian (normal) distribution. An approach to create a simple model is to assume that the data is described bt a Gaussian distribution with no co-variance(indeoendent dimensions) between dimensions.

```python
model_gnbc = MultinomialNB(alpha = 0.6)
model_gnbc.fit(X_train.toarray(), y_train)

predict_gnbc = model_gnbc.predict(X_test.toarray())

print(f"Classification Report : \n{B}{Red}{classification_report(y_true = y_test, y_pred = predict_gnbc)}{R}")
print("- " * 30)
print(f"Train Score  : {B}{Blue}{model_gnbc.score(X_train.toarray(), y_train)}{R}")
print(f"Test Score   : {B}{Blue}{model_gnbc.score(X_test.toarray(), y_test)}{R}")
```

### Confusion Matrix

```python
fig, axes = plt.subplots(1, 2, figsize = (15, 8))

sns.heatmap(
    confusion_matrix(y_true = y_test, y_pred = predict_mnbc),
    ax = axes[0],
    annot = True,
    fmt = 'd',
    annot_kws={"size": 20},
    cmap = 'Blues',
    cbar = False
)
axes[0].set_title("Multinomial Naive Bayes Classification\n", fontweight = 'bold', fontsize = 18)
axes[0].set_xlabel("Predicted Label")
axes[0].set_ylabel("True Label")

sns.heatmap(
    confusion_matrix(y_true = y_test, y_pred = predict_gnbc),
    ax = axes[1],
    annot = True,
    fmt = 'd',
    annot_kws={"size": 20},
    cmap = 'Greens',
    cbar = False
)
axes[1].set_title("Gaussian Naive Bayes Classification\n", fontweight = 'bold', fontsize = 18)
axes[1].set_xlabel("Predicted Label")
axes[1].set_ylabel("True Label")

plt.tight_layout()
plt.show()
```
